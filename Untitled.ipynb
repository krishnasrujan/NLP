```python
import pandas as pd
import numpy as np
import re
import string
from bs4 import BeautifulSoup as bs
from collections import Counter
import cv2
import matplotlib.pyplot as plt
IMG_SIZE=128
```

```python
X=np.load("img_data.npy",allow_pickle=True)
X = X.reshape(-1,IMG_SIZE,IMG_SIZE,3)
X.shape
```

```python
y = np.load('encoded_target.npy',allow_pickle=True)
y.shape
```

```python
X = X/255
```

```python
plt.imshow(X[0])
```

```python
plt.bar(height = dict(Counter(y)).values(),x = [0,1,2])
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Flatten,Activation,Dense,Conv2D,MaxPooling2D,Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras import regularizers
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from tensorflow.keras.applications import vgg16
from tensorflow import keras
from tensorflow.keras.models import Model

from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
train_datagen = ImageDataGenerator(zoom_range=0.1, rotation_range=30,
                                   width_shift_range=0.1, height_shift_range=0.1, shear_range=0.2, 
                                   horizontal_flip=True, fill_mode='nearest')

val_datagen = ImageDataGenerator(zoom_range=0.1, rotation_range=30)
```

```python
train_generator = train_datagen.flow(x_train, y_train, batch_size=32)
val_generator = val_datagen.flow(x_test, y_test, batch_size=32)
```

```python
resnet50 = keras.applications.ResNet50(include_top=False, weights='imagenet', 
                                     input_shape=(IMG_SIZE,IMG_SIZE,3))

output = resnet50.layers[-1].output
output = tf.keras.layers.Flatten()(output)
resnet50_model = Model(resnet50.input, output)

resnet50_model.trainable = False
for layer in resnet50_model.layers:
    layer.trainable = False
```

```python
from tensorflow.keras import optimizers

model = Sequential()
model.add(resnet50_model)
model.add(Dense(512, activation='relu', input_dim=(IMG_SIZE,IMG_SIZE,3)))
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(3, activation='softmax'))

opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='sparse_categorical_crossentropy',optimizer=opt,metrics=['accuracy'])
```

```python
class_weight = {0: 2.,
                1: 1.,
                2: 2.}
```

```python
#before giving class weight train acc was 81 and test acc was 88
```

```python
history = model.fit_generator(train_generator, epochs=5, validation_data=val_generator, verbose=1,class_weight=class_weight)
```
